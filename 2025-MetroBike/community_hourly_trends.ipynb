```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# merged

```python
cm_merged = pd.read_csv("cm_merged.csv")
cm_grouped = cm_merged.groupby(['Checkout Hour', 'Community_x']).size().reset_index(name='Count')
cm_grouped
```

```python
custom_palette = {
    0: (30/255, 120/255, 230/255),   # blue
    1: (50/255, 160/255, 70/255),    # green
    2: (255/255, 180/255, 40/255),   # orange
    3: (240/255, 100/255, 100/255),  # pink
    4: '#CCA3E1' # purple
}

plt.figure(figsize=(12, 6))

for community in sorted(cm_grouped['Community_x'].unique()):
    subset = cm_grouped[cm_grouped['Community_x'] == community]
    plt.plot(
        subset['Checkout Hour'],
        subset['Count'],
        marker='o',
        linestyle='-',
        color=custom_palette.get(community, 'gray'),
        label=f'{community}'
    )

plt.xlabel('Checkout hour')
plt.ylabel('Usage count')
plt.xticks(range(0, 24))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Community')
plt.tight_layout()

plt.savefig("cm_merged.png", dpi=300)

plt.show()
```
