```python
import pandas as pd
```

```python
df = pd.read_csv("D:/heat/MTA_Subway_Origin-Destination_Ridership_Estimate__2024_20250627.csv")
```

```python
# Timestamp is datetime type
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

# Min and Max
min_time = df["Timestamp"].min()
max_time = df["Timestamp"].max()

print(f"ðŸ“… Earliest timestamp: {min_time}")
print(f"ðŸ“… Latest timestamp: {max_time}")
```

```python
df["Timestamp"].unique()
```

```python
# Unique values ì¶”ì¶œ í›„ ì •ë ¬
unique_timestamps = pd.DataFrame(sorted(df["Timestamp"].unique()), columns=["Unique Timestamps"])
unique_timestamps.to_csv("unique_timestamps.csv", index=False)
```

```python
# Filter: June to August 2024
df_summer = df[(df["Timestamp"] >= "2024-06-01") & (df["Timestamp"] < "2024-09-01")]
df_summer
```

```python
# Sorting
df_sorted = df_summer.sort_values("Timestamp").reset_index(drop=True)
df_sorted
```

```python
df_sorted.to_csv("nyc_subway_summer.csv", index=False)
```
